```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
sales_data={'Electronics':45000,'Clothing':30000,'Grocery':25000,'Mobile':20000}
category=list(sales_data.keys())
actual_sale=list(sales_data.values()) # ---> data
print(category)
print(actual_sale)
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(10,5))
x=np.random.random(10)
ax1.plot(x,x/2)
ax2.scatter(np.random.random(10),np.random.random(10))
ax3.bar(category,actual_sale)
ax4.hist(np.random.randn(1000))
```

## Subplots using indexing

```python
our_lines=np.linspace(0,10,100)
```

```python
product_prices={'Hair oil':20, 'Facewash': 30, 'Lipbalm':10, 'Face cream':15}
```

```python
our_numbers=np.random.random(10)
```

```python
our_dots=np.random.randn(1000)
```

```python
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(8,5))
ax[0,0].plot(our_lines,our_lines/2)
ax[0,1].bar(list(product_prices.keys()),list(product_prices.values()),color=['pink','red','salmon','green'])
ax[1,0].scatter(np.random.randn(100),np.random.randn(100),color='orange')
ax[1,1].hist(our_dots,color='red')
plt.tight_layout()
```

```python
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(8,5))

#first row (0),first column(0) --> 0,0 ; second row(1), second column (1) --> 1,1

ax[0,0].plot(our_lines,our_lines/2) # --> plot 1
ax[0,1].bar(list(product_prices.keys()),list(product_prices.values()),color=['pink','red','salmon','green'])
ax[1,0].scatter(np.random.randn(100),np.random.randn(100),color='orange')
ax[1,1].hist(our_dots,color='red')

ax[0,0].set_title('Line Plot Example') #--> plot 1
ax[0,0].set_xlabel('X Axis')
ax[0,0].set_ylabel('Y Axis')
ax[0,0].grid(True,linestyle='--',alpha=0.7, color='lightblue')

ax[0,1].set_title('Product Sales Example') #--> plot 2
ax[0,1].set_xlabel('Product Name')
ax[0,1].set_ylabel('Product Sale')
ax[0,1].grid(True,linestyle='--',alpha=0.7, color='lightblue', axis='y')

ax[1,0].set_title('Scatter Plot') #--> plot 3
ax[1,0].set_xlabel('Product Name')
ax[1,0].set_ylabel('Product Sale')
ax[1,0].grid(True,linestyle='--',alpha=0.7, color='lightblue')

ax[1,1].set_title('Histogram Example') # --> plot 4
ax[1,1].set_xlabel('Value Range')
ax[1,1].set_ylabel('Frequency')
ax[1,1].grid(True,linestyle='--',alpha=0.7, color='lightblue')

fig.suptitle('Matplotlib Subplots example using 2 rows and 2 cols')
plt.tight_layout()
plt.savefig('subplotex.jpeg',dpi=300,bbox_inches='tight')
```
