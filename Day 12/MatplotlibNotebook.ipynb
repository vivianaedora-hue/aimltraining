# First Example to create a line chart from our data
    ## Students.xlsx for fetching data
    ## Generate line chart based on Students name and Student marks

```python
import pandas as pd
```

```python
df=pd.read_excel('C:/Users/Asus/Desktop/aimltraining/Day 10/students.xlsx')
print(df)
```

```python
names=df['Name']
marks=df['Marks']
```

```python
plt.plot(names,marks,marker='s',color='green')
plt.title('Student Marks Line Chart')
plt.xlabel('Student Name')
plt.ylabel('Students Marks')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(names,marks,marker='o',color='orange')
plt.title('Student Progress Chart')
plt.xlabel('Student Name')
plt.ylabel('Students Marks')
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.plot(names,marks,marker='o',color='orange')
plt.title('Student Progress Chart')
plt.xlabel('--Name--')
plt.ylabel('--Marks--')
plt.grid(True)
plt.show()
```

```python
print('Dashed Line')
plt.plot(names,marks,marker='o',color='orange',linestyle='--')
plt.show()
```

```python
print('Dashed Line')
plt.plot(names,marks,marker='o',color='orange',linestyle=':')
plt.show()
```
