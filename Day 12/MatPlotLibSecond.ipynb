```python
players=['MSD','Rohit','Virat','Gill','Shami']
match1_score=[45,67,56,78,89]
match2_score=[55,78,40,80,97]
match3_score=[90,40,20,30,60]
plt.figure(figsize=(8,5))

plt.scatter(players, match1_score, color='red', label='Match One', s=100)
plt.scatter(players, match2_score, color='green', label='Match Two', s=100, marker='d')
plt.scatter(players, match3_score, color='blue', label='Match Three', s=150, marker='p')

plt.title('Comparison of Players in Two Matches')
plt.xlabel('Players')
plt.ylabel('Score')
plt.legend(loc='upper right') # 'lower right'

plt.grid(True, linestyle='--', alpha=0.5)

plt.show()
```

## Write a comparison that shows average calories burned for each workout type from final_dataset

```python
df=pd.read_csv('Final_data.csv')

# 1. Group by workout type and find the average calories burned
avg_calories = df.groupby('Workout_Type')['Calories_Burned'].mean()

# 2. Print results
print("Average Calories Burned for Each Workout Type:")
print(avg_calories)

# 3. Visualize
plt.scatter(avg_calories.index, avg_calories.values, color='skyblue', s=100,marker='p')

plt.title('Average Calories Burned per Workout Type')
plt.xlabel('Workout Type')
plt.ylabel('Average Calories Burned')
plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```

```python
df=pd.read_csv('Final_data.csv')
avg_cal=df.groupby('Workout_Type')['Calories_Burned'].mean()
selected_index=avg_cal.index
selected_value=avg_cal.values
plt.scatter(selected_index,selected_value,color='green', s=100,marker='p')

plt.title('Average Calories Burned per Workout Type')
plt.xlabel('Workout Type')
plt.ylabel('Average Calories Burned')
plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```
