```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
temp=np.random.randint(20,100,size=50)

plt.scatter(x,y,c=temp,cmap='viridis',s=120, edgecolor='black')

plt.colorbar(label='Temperature')
plt.title('Scatter plot Colored by temperature')
plt.xlabel('X-random')
plt.ylabel('Y-Random')
plt.show()
```

```python
height_cm=[150,155,160,165,170,175,180,185]
weight_kg=[50,53,56,59,62,66,68,72]
age_years=[20,22,25,28,30,35,38,40]
```

```python
plt.figure(figsize=(8,5))
```

```python
plt.scatter(height_cm, weight_kg, c=age_years, cmap='plasma', s=120, alpha=0.9)
plt.colorbar(label='Age (years)')

plt.title('Scatter plot: Height vs Weight (colored by age)')
plt.xlabel('Height(cm)')
plt.ylabel('Weight(kg)')

plt.grid(color='blue', linestyle='--', linewidth=0.7)
plt.show()
```

# Pie Plot Example
    ## plt.pie(x,labels='',colors='',autoopct='',startangle='', shadow=, explode=, radius)

```python
depts=['HR','Web-development','App-development','Agile-scrum','Testing']
expenses=[30,10,30,15,15]
```

```python
plt.figure(figsize=(7,7))
```

```python
plt.pie(expenses,labels=depts, autopct='%1.1f%%',startangle=140,explode=(0,0.05,0,0,0), shadow=True)
plt.title('Department wise expense report')
plt.show()
```

```python
plt.pie(expenses,labels=depts, colors=plt.cm.rainbow([0.1,0.3,0.5,0.7,0.9]), autopct='%1.1f%%',startangle=140,explode=(0,0.05,0,0,0), shadow=True)
#color magma,rainbow,inferno
plt.title('Department wise expense report')
plt.show()
```

# create a pie plot using following data
## sales_data={'Mobile':45000,'Clothing':30000,'Groceries':35000,'Home decor':20000}
### show percentage with two decimal
### use colormap to give the diffferent color to your pie

```python
category=['Mobile','Clothing','Groceries','Home decor']
sales=[45000,30000,35000,20000]
```

```python
plt.figure(figsize=(7,7))
```

```python
plt.pie(sales,labels=category, autopct='%1.2f%%',startangle=140,explode=(0,0.05,0,0), shadow=True)
plt.title('Sales by category Report')
plt.show()
```

```python
sales_data={'Mobile':45000,'Clothing':30000,'Groceries':35000,'Home decor':20000}
cat=list(sales_data.keys())
sale=list(sales_data.values())

plt.figure(figsize=(7,7))

plt.pie(sale,labels=cat, colors=plt.cm.rainbow([0.1,0.3,0.5,0.9]), textprops={'color':'brown','fontsize':10,'fontweight':'bold'},
        autopct='%1.2f%%',startangle=140,explode=(0,0.05,0,0), shadow=True)
plt.title('Sales by category Report'),
#plt.savefig('filenamewtihextension',dpi=value,bbox_inches='')
plt.savefig('salesbycategoryreport.png',dpi=300,bbox_inches='tight')
plt.show()
```

## Histogram example


```python
marks=[45,60,80,55,62,90,89,50,56,60,80,55,62,90,89,58,90,56,67,83]
plt.style.use('ggplot')
plt.figure(figsize=(8,5))
plt.hist(marks,bins=5,color='skyblue', edgecolor='black', alpha=0.8)

plt.title('Histogram for Students Marks')
plt.xlabel('Marks Range')
plt.ylabel('Number of Students')

plt.grid(color='blue', linestyle='--', linewidth=0.7)
plt.show()
```
