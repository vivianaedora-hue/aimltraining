```python
import matplotlib.pyplot as plt
```

```python
months=['Jan','Feb','Mar','Apr']
sales=[2000,3000,2500,3600]
```

## plt.plot(x,y,color='color name', linestyle='line_style', linewidth=value, marker='marker symbol', label='label name')

```python
plt.plot(months,sales,color='hotpink', linestyle='--', linewidth=2, marker='s', label='Monthly Sales Data')
plt.xlabel('Months')
plt.ylabel('Sales per month')
plt.title('Monthly Sales Report')
plt.show()
```

```python
plt.plot(months,sales,color='hotpink', linestyle='--', linewidth=2, marker='s', label='Monthly Sales Data')
plt.xlabel('Months')
plt.ylabel('Sales per month')
plt.title('Monthly Sales Report')

plt.xlim(0,4)
plt.ylim(2000,4000)
plt.grid(True)
plt.legend(loc='lower right')

plt.show()
```

```python
import numpy as np
```

```python
x=np.linspace(0,2,100)
plt.plot(x,x,label='Linear')
plt.plot(x,x**2,label='Quad')
plt.plot(x,x**3,label='Cubic')

plt.xlabel('X lines')
plt.ylabel('Y lines')

plt.title('Simple lines plot')
plt.legend(loc='upper left')
plt.show()
```

# Bar plot example
    ## plt.bar(x,y,color='',edgecolor='')

```python
categories=['Electronics','Mobile','Books','Groceries','Home Decoration']
sales=[50000,60000,30000,90000,40000]
bar_colors=['red','blue','green','purple','orange']
plt.figure(figsize=(8,5))
plt.bar(categories,sales,color=bar_colors, edgecolor='black')

plt.xlabel('Categories')
plt.ylabel('Sales')
plt.title('Categories wise per sales report')

plt.grid(axis='y',linestyle='--')
plt.tight_layout
plt.show()
```

```python
categories=['Electronics','Mobile','Books','Groceries','Home Decoration']
sales=[50000,60000,30000,90000,40000]
#bar_colors=['red','blue','green','purple','orange']
plt.figure(figsize=(8,5))
plt.bar(categories,sales,color='blue', edgecolor='black')

plt.xlabel('Categories')
plt.ylabel('Sales')
plt.title('Categories wise per sales report')

plt.grid(axis='y',linestyle='--',color='green',alpha=0.5)
plt.xticks(rotation=30, fontsize=12, color='magenta')
plt.yticks([20000,40000,60000,80000,100000],['20k','40k','60k','80k','1M'], fontsize=12, color='green')

plt.legend(title='Legend')
plt.tight_layout
plt.show()
```

# Scatter plot
    ## plt.scatter(x,y,s='', c='colorName', marker='makername', alpha='transp.', edgecolor='colorname', label='lable of legend')

```python
height_cm=[150,155,160,165,170,175,180,185]
weight_kg=[50,53,56,59,62,66,68,72]

plt.figure(figsize=(8,5))
plt.scatter(height_cm, weight_kg, s=100, c='lightgreen', marker='*', alpha=0.9, edgecolor='black', label='weight/height')

plt.title('Scatter plot height vs weight')
plt.xlabel('Height(cm)')
plt.ylabel('Weight(kg)')

plt.xlim(140,190)
plt.ylim(40,90)

plt.xticks(rotation=45, fontsize=12, color='green')
plt.yticks([50,55,60,65,70,75,80,85], ['50kg','55kg','60kg','65kg','70kg','75kg','80kg','85kg'], fontsize=12, color='green')

plt.grid(color='lightblue', alpha=0.5, linestyle='--', linewidth=0.5)
plt.legend()
plt.show()
```
