```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('netflix_data.csv')
print(df.head())
```

```python
df=df.dropna(subset=['release_year'])
```

```python
df['release_year']=df['release_year'].astype(int)
```

```python
plt.figure(figsize=(5,3))
plt.hist(df['release_year'],bins=100,color='skyblue', edgecolor='black', alpha=0.8)

plt.title('Netflix Released Title by Year')
plt.xlabel('Release year')
plt.ylabel('Number of Titles')

plt.grid(color='blue', linestyle='--', linewidth=0.7)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('netflix_data.csv')
df=df.dropna(subset=['release_year'])
df['release_year']=df['release_year'].astype(int)

plt.figure(figsize=(8,5))
plt.hist(df['release_year'],bins=100,color='skyblue', edgecolor='black', alpha=0.8)

plt.title('Netflix Released Title by Year')
plt.xlabel('Release year')
plt.ylabel('Number of Titles')

plt.grid(color='blue', linestyle='--', linewidth=0.7)
plt.tight_layout()
plt.savefig('netflix_bar.pdf',dpi=300, bbox_inches='tight')
plt.show()
```

```python
df=df.dropna(subset=['release_year','type'])
```

```python
df['release_year']=df['release_year'].astype(int)
```

```python
movies=df[df['type']=='Movie']
shows=df[df['type']=='TV Show']
```

```python
plt.figure(figsize=(10,5))
plt.hist([movies['release_year'], shows['release_year']],bins=30,color=['pink','skyblue'], edgecolor='black', alpha=0.8, label=['Movies','TV Shows'])

plt.title('Netflix Released Title by Year')
plt.xlabel('Release year')
plt.ylabel('Number of Titles Released')

plt.grid(axis='y', color='blue', linestyle='--', linewidth=0.7)
plt.tight_layout()
plt.legend()
plt.savefig('netflix_showvsmovies.jpg',dpi=300, bbox_inches='tight')
plt.show()
```

# Use netflix_data.csv dataset
### print how many Movies and TV shows are available for Netflix
### Display Movies and Tv shows proportion in a pie plot
### Save file as pdf


```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('netflix_data.csv')

type_counts = df['type'].value_counts() # --> # 2. Check unique types available (Movie / TV Show
print("Number of Movies and TV Shows on Netflix:")
print(type_counts)
```

```python
plt.figure(figsize=(8,4))
#colors = plt.cm.Pastel1(range(len(type_counts)))  # Colormap for variety

plt.bar(
    type_counts.index,
    type_counts.values,
    color=['pink','skyblue'],
    edgecolor='black',
    alpha=0.8
   )

plt.title('Netflix Movies and TV Shows Distribution')
plt.xlabel('Types')
plt.ylabel('Counts')

plt.grid(axis='y', color='blue', linestyle='--', linewidth=0.7)
plt.tight_layout()
plt.legend()
# 4. Save the figure as a PDF file
plt.savefig('netflix_movies_and_tvshows_distribution.pdf', dpi=300, bbox_inches='tight')

# 5. Display the pie chart
plt.show()
```

```python
plt.figure(figsize=(8,4))
#colors = plt.cm.Pastel1(range(len(type_counts)))  # Colormap for variety

plt.pie(type_counts,
    labels=type_counts.index,
    colors=['pink','skyblue'],
    autopct='%1.2f%%',
    startangle=140,
    shadow=True,
    explode=(0.05,0.05))

plt.title('Netflix Movies and TV Shows Distribution')

#plt.tight_layout()
plt.legend(title='Movies and Shows')
# 4. Save the figure as a PDF file
plt.savefig('netflix_pieplot.pdf', dpi=300, bbox_inches='tight')

# 5. Display the pie chart
plt.show()
```
